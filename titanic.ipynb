```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.model_selection import cross_val_score

y = train['Survived']
features = ['Pclass', 'Sex', 'SibSp', 'Parch'] 
X = pd.concat([pd.get_dummies(train[features], drop_first=True), train['Age']], axis=1)
X_test = pd.concat([pd.get_dummies(test[features], drop_first=True), test['Age']], axis=1)

#model = HistGradientBoostingClassifier(learning_rate=0.1, max_iter=300, max_depth=5, random_state=1)
model = HistGradientBoostingClassifier(learning_rate=0.05, max_iter=500, max_depth=5, random_state=1)
print(cross_val_score(model, X, y, cv=5))


model.fit(X, y)
print(classification_report(y, model.predict(X)))
predictions = model.predict(X_test)

output_HXGB = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
```

# Save predictions

```python
output_HXGB.to_csv('submission.csv', index=False)
```
